Linea Diagonal dentro de cuadrado negro

```python
# Se crea una matriz tridimensional de tamaño W y H
#  el tercer parámetro 3 representa los tres canales de color
# (rojo, verde y azul).
im1 = np.zeros((W,H, 3), np.uint8)

#Se utiliza un bucle for para iterar sobre los valores de i
#desde 0 hasta W-1. En cada iteración, se establece el valor
#de los píxeles en la posición (i, i) en la imagen im1 en verde
#(0, 255, 0). Esto crea una línea diagonal verde que atraviesa
#la imagen desde la esquina superior izquierda hasta la esquina inferior derecha.
for i in range(W): im1[i,i] = (0,255,0)

cv2_imshow(im1)
```

```python
# Se inicializan dos variables con valor de 300 cada una
# una representa la altura y el ancho de la iamgen respectivamente.
h, w = 300, 300

#se crea una matriz tridimensional, donde 3 representa los tres cabales de color
# red, green y blue. todas los elementos de la matriz se inicializan a 0
# creando asi una imagen negra
image = np.zeros((h, w, 3), np.uint8)

#se calcula la posicion inicial, para dibujar un cuadrado azul
# la formula (h - 90) // 2 y calucla el punto central verticalmente
# y lo mismo para el punto central horizontal
row = (h - 90) // 2
col = (w - 90) // 2

#en este ciclo se usa para iterar sobre los pixeles de un cudarado
# de 90 *90 en la imagen y se establece el valor de los pixeles en azul
for i in range (90):
  for j in range (90):
    image[row + i, col + j] = (0, 0, 255)

row2 = (h - 70) // 2
col2 = (w - 70) // 2
for i in range (70):
  for j in range (70):
    image[row2 + i, col2 + j] = (0, 255, 0)

row3 = (h - 50) // 2
col3 = (w - 50) // 2
for i in range (50):
  for j in range (50):
    image[row3 + i, col3 + j] = (255, 0, 0)

row4 = (h - 30) // 2
col4 = (w - 30) // 2
for i in range (30):
  for j in range (30):
    image[row4 + i, col4 + j] = (0, 0, 0)

#finalmente se muestra la imagen resultante utilzando la siguiente funcion
# de la biblioteca openCv
cv2_imshow(image)
```

```python
import numpy as np
from google.colab.patches import cv2_imshow

def generar_ondas(h, w, num_waves, amplitude, frequency, color):
    """
    Genera ondas horizontales de un color en una imagen negra.

    Parámetros:
        h (int): Altura de la imagen.
        w (int): Ancho de la imagen.
        num_waves (int): Número de ondas.
        amplitude (int): Amplitud de las ondas.
        frequency (int): Frecuencia de las ondas.
        color (tuple): Color en formato (B, G, R) para las ondas.

    Retorna:
        numpy.ndarray: Imagen con ondas horizontales.
    """
    image = np.zeros((h, w, 3), np.uint8)

    for _ in range(num_waves):
        for y in range(h):
            for x in range(w):
                wave_position = x / w * 2 * np.pi * frequency * num_waves
                wave_amplitude = np.sin(wave_position) * amplitude
                if y > (h // 2) + wave_amplitude and y < (h // 2) + amplitude:
                    image[y, x] = color

    return image

# Parámetros
h, w = 300, 300
num_waves = 5  # Número de ondas
amplitude = 10  # Amplitud de las ondas
frequency = 20  # Frecuencia de las ondas
color = (0, 255, 255)  # Color en formato (B, G, R)

# Generar imagen con ondas
image_with_waves = generar_ondas(h, w, num_waves, amplitude, frequency, color)

# Mostrar la imagen
cv2_imshow(image_with_waves)
```

```python
import cv2
from google.colab.patches import cv2_imshow
import numpy as np

W = 300
H = 300

im = np.zeros((H, W), np.uint8)

# Definir el número de líneas horizontales (y su frecuencia) basadas en la función seno
num_lines = 10
frequency = 20  # Frecuencia de la onda seno

# Dibujar las líneas horizontales que siguen una función seno
for i in range(num_lines):
    amplitude = int(H / (50))  # La amplitud de la línea varía con la frecuencia
    y_offset = H // (num_lines + 1) * (i + 1)

    # Generar puntos basados en la función seno
    x = np.arange(0, W)
    y = amplitude * np.sin(2 * np.pi * frequency * x / W) + y_offset

    # Convertir los puntos a enteros
    points = np.column_stack((x, y)).astype(np.int32)

    # Dibujar la línea
    cv2.polylines(im, [points], isClosed=False, color=255, thickness=1)

# Mostrar la imagen
cv2_imshow(im)
```
